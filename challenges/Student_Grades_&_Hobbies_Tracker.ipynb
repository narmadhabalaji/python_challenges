```python
# Creating Dictionary
student_info={
    "Asha": 85,
    "Rahul": 78,
    "Meera": 92
}
print(student_info)


## Allowing the user to add or update grades for at least 3 students.

for i in range(3):
  name = input('\nEnter the student name:')
  grade = int(input('Enter the student grade:'))
  student_info[name]=grade

print('\nUpdated student info:',student_info)

## Creating a set of hobbies and allowing the user to input at least 3 hobbies
hobbies=set()

count = int(input("\nHow many hobbies do you want to enter? "))

for i in range(count):
    hobby = input(f"Enter hobby {i+1}: ")
    hobbies.add(hobby)

#Converting the set of hobbies into a frozenset and displaying it.
frozen_hobbies=frozenset(hobbies)
print(frozen_hobbies)
```
