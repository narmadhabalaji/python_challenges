```python
product_categories=('Electronics', 'Stationery', 'Clothing')
print(product_categories)

# **List of Products**
products=['Laptop', 'Pen', 'Notebook', 'Tshirt']
print(products)

# New product from User
new_product=input('Enter a new product: ')
products.append(new_product)

print('Updated product list: ',products)
# Item to be removed
prod_ind=int(input('Enter the index no of the product to be removed (starting index is 0):'))
products.pop(prod_ind)

# Final list of products and the product categories
print("\nFinal Inventory Details:")
print("Product Categories:", product_categories)
print("Products:", products)
```
