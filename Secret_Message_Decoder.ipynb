```python
secret_message = input('Enter your secret message: ')
print('\nLength of secret msg: ',len(secret_message))
upper_txt=secret_message.upper()
lower_txt=secret_message.lower()
print('\nSecret msg in UPPER CASE:',upper_txt, '\n\nSecret msg in lower case:', lower_txt)
print('\nReplaced space to underscore:', upper_txt.replace(" ","_"))
print('\nEvery 3rd character of secret msg:',upper_txt[::3])
print("\nReversed the message: ",upper_txt[::-1])
print("\nSecret message is alphabet: ",secret_message.isalpha())
```
